```python
#import packages..
from numpy import array 
from matplotlib import pyplot 
from statsmodels.stats.power import TTestIndPower 

# input levers for power analysis

effect_sizes = array([0.2, 0.5, 0.8])
alpha = array([0.05, 0.01])
sample_sizes = array(range(2, 100,30))

# Lets plot power Graph by changing Levers.. for varying Samples and effect sizes based on cohen's d 
analysis = TTestIndPower()
analysis.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes,alpha=0.05 ,title='Power Analysis For Different Effect Size' + '\n' + r'$\alpha = 0.05$')
pyplot.show()

analysis.plot_power(dep_var='effect_size', nobs=sample_sizes, effect_size=effect_sizes,alpha=0.05,title='Power Analysis For Different Sample Size' + '\n' + r'$\alpha = 0.05$')
pyplot.show()

analysis = TTestIndPower()
analysis.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes,alpha=0.01 ,title='Power Analysis For Different Effect Size' + '\n' + r'$\alpha = 0.01$')
pyplot.show()

analysis.plot_power(dep_var='effect_size', nobs=sample_sizes, effect_size=effect_sizes,alpha=0.01,title='Power Analysis For Different Sample Size' + '\n' + r'$\alpha = 0.01$')
pyplot.show()

```
